```python
import matplotlib.pyplot as plt
import numpy as np

schedule = [
    (23, 24, "sleep", "lightblue"),
    (0, 8, "sleep", "lightblue"),
    (8, 9, "breakfast", "plum"),
    (9, 10, "break", "gold"),
    (10, 11, "volleyball", "yellow"),
    (11, 12, "study", "khaki"),
    (12, 13, "lunch", "lightcoral"),
    (13, 14, "break", "pink"),
    (14, 15, "study", "lavender"),
    (15, 16, "walk", "lightblue"),
    (16, 17, "saxophone", "orange"),
    (17, 18, "shower", "lightblue"),
    (18, 20, "dinner", "plum"),
    (20, 23, "break", "lightpink"),
]

fig, ax = plt.subplots(figsize=(5, 5), subplot_kw={'projection': 'polar'})

for start, end, label, color in schedule:
    theta_start = (start / 24) * 2 * np.pi
    theta_end = (end / 24) * 2 * np.pi
    width = theta_end - theta_start

    ax.bar(
        x=theta_start + width / 2,
        height=1,
        width=width,
        bottom=0,
        color=color,
        edgecolor='white',
        linewidth=2,
        align='center',
        alpha=0.85
    )

    #     
    angle = (theta_start + theta_end) / 2
    ax.text(angle, 0.85, label,
            rotation=0,
            ha='center', va='center', fontsize=10, color='black')


ax.set_theta_direction(-1)
ax.set_theta_offset(np.pi / 2)

ax.set_yticklabels([])
ax.set_xticks(np.linspace(0, 2 * np.pi, 24, endpoint=False))
ax.set_xticklabels([f"{i}" for i in range(0, 24)])
ax.grid(False)

plt.title("sara's time table", fontsize=15)
plt.show()
```
